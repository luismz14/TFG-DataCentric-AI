```python
from pathlib import Path

import pandas as pd

TEST_CSV = Path("data/test/external_test.csv")
TEST_IMAGES_DIR = Path("data/test/external_test")
TEST_CROPPED_IMAGES_DIR = Path("data/test/images_cropped")
TEST_EXISTING_CSV = Path("data/test/external_test_existing.csv")

test_df = pd.read_csv(TEST_CSV)
existing_mask = test_df["filename"].astype(str).map(lambda filename: (TEST_IMAGES_DIR / filename).exists())
print(f"Rows in test CSV: {len(test_df)}")
print(f"Images found: {int(existing_mask.sum())}")
print(f"Rows without image: {int((~existing_mask).sum())}")

test_df.loc[existing_mask].reset_index(drop=True).to_csv(TEST_EXISTING_CSV, index=False)

test_cropped_df = create_cropped_dataset(
    source_csv=TEST_EXISTING_CSV,
    source_images_dir=TEST_IMAGES_DIR,
    cropped_images_dir=TEST_CROPPED_IMAGES_DIR,
    output_csv=TEST_CSV,
)
TEST_EXISTING_CSV.unlink(missing_ok=True)

test_cropped_df[["filename", "histology"]].head()
```
