```python
import pandas as pd
```

```python
path = r"../data/master_dataset.csv"

df = pd.read_csv(path, delimiter=";")
df.head()
```

```python
df.rename(columns={'Unnamed: 0': 'photoFileName'}, inplace=True)

# The target column is "assumed histology", so let's see all the differents classes it has and the count of each.
print(df["assumed histology"].value_counts())
```

```python
df["assumed histology"] = df["assumed histology"].str.replace(" Hyperplastic", "Hyperplastic")
df["assumed histology"] = df["assumed histology"].str.replace("Sessile serrated adenoma (with displasia)", "Sessile serrated adenoma")
print(df["assumed histology"].value_counts())
```
